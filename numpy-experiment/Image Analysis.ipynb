```python
import cv2
dir(cv2)
```

```python
img_grayscale = cv2.imread("smallgray.png", 0)
print(img_grayscale)
```

```python
img_bgr = cv2.imread("smallgray.png", 1)
print(img_bgr)
```

```python
cv2.imwrite("new_smallgray.png", img_grayscale)
```

```python
print("Indexing:\n")
print(img_grayscale[1:2, 2:4])
print(img_grayscale[1:, 2:])
print(img_grayscale[:, 2:4])
print(img_grayscale[1:2, :])
print(img_grayscale[2, 0])
```

```python
print("Iteration:\n")
print("Row Iteration:\n")
for i in img_grayscale:
    print(i)
print("\nColumn Iteration:\n")
for i in img_grayscale.T:
    print(i)
print("\nIndividual Iteration:\n")
for i in img_grayscale.flat:
    print(i)
```

```python
ims_h = numpy.hstack((img_grayscale,img_grayscale,img_grayscale))
print(ims_h)
```

```python
ims_v = numpy.vstack((img_grayscale,img_grayscale,img_grayscale))
print(ims_v)
```

```python
lst_h = numpy.hsplit(ims_h, 3)
print(lst_h)
```

```python
lst_v = numpy.vsplit(ims_v, 3)
print(lst_v)
```
